Que.1.How to implement a simple text classification model using LSTM in Keras0


```python
!pip install tensorflow numpy pandas scikit-learn

#step 2
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load IMDB dataset
max_features = 10000  # Number of words to consider
maxlen = 500  # Maximum length of each review

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Pad sequences to ensure uniform input size
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

#Define LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

model = Sequential()
model.add(Embedding(input_dim=max_features, output_dim=128, input_length=maxlen))  # Embedding layer
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))  # LSTM layer with dropout
model.add(Dense(1, activation='sigmoid'))  # Output layer (binary classification)

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))
#Evaluate the model
score, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
#make Predection
predictions = model.predict(x_test)
```

Que.2.9G How to generate sequences of text using a Recurrent Neural Network (RNN)


```python
!pip install tensorflow numpy pandas
#prepare text data
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import get_file

# Example: Using a small text corpus (you can use a larger dataset for better results)
text = "This is a simple example of text generation using RNN."

# Preprocessing: Convert the text into integer sequences
chars = sorted(set(text))  # Get all unique characters
char_to_idx = {ch: idx for idx, ch in enumerate(chars)}  # Character to index mapping
idx_to_char = {idx: ch for ch, idx in char_to_idx.items()}  # Index to character mapping

# Convert text to integer sequence
input_sequence = [char_to_idx[ch] for ch in text]

# Create input-output pairs for training (sliding window approach)
sequence_length = 10  # Define the sequence length for input data
X = []
y = []
for i in range(len(input_sequence) - sequence_length):
    X.append(input_sequence[i:i+sequence_length])
    y.append(input_sequence[i+sequence_length])

X = np.array(X)
y = np.array(y)

#Step 3: Define the RNN Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding

model = Sequential()

# Embedding layer: Convert integer sequences into dense vectors
model.add(Embedding(input_dim=len(chars), output_dim=10, input_length=sequence_length))

# RNN layer: Basic RNN for sequence modeling
model.add(SimpleRNN(128, activation='relu', return_sequences=False))

# Output layer: One neuron for each character in the vocabulary (softmax activation)
model.add(Dense(len(chars), activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


#Step 4: Train the Model

model.fit(X, y, epochs=200, batch_size=32)

#Step 5: Text Generation Function

def generate_text(model, seed, num_chars=100):
    # Convert the seed text to integers
    seed_input = [char_to_idx[ch] for ch in seed]

    generated_text = seed
    model.reset_states()

    for _ in range(num_chars):
        # Convert seed input to shape (1, sequence_length)
        input_sequence = np.expand_dims(seed_input, axis=0)

        # Predict the next character
        predicted_prob = model.predict(input_sequence)
        predicted_idx = np.argmax(predicted_prob, axis=-1)[0]

        # Append the predicted character to the generated text
        predicted_char = idx_to_char[predicted_idx]
        generated_text += predicted_char

        # Update seed for the next iteration
        seed_input.append(predicted_idx)
        seed_input = seed_input[1:]  # Keep the sequence length fixed

    return generated_text

# Generate text starting with a seed string
seed_text = "This is"
generated_text = generate_text(model, seed_text, num_chars=100)
print(generated_text)
```

Que.3.How to perform sentiment analysis using a simple CNN model0

```python


import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential # Import Sequential from tensorflow.keras
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout # Import layers from tensorflow.keras
from tensorflow.keras.preprocessing.text import Tokenizer # Import Tokenizer from tensorflow.keras.preprocessing.text
from tensorflow.keras.preprocessing.sequence import pad_sequences # Import pad_sequences from tensorflow.keras.preprocessing.sequence
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


#step 2- Load dataset

# Example dataset
data = {'text': ['I love this movie', 'This is a bad movie', 'Amazing story and acting', 'Not worth the watch'],
        'sentiment': [1, 0, 1, 0]}  # 1: Positive, 0: Negative
df = pd.DataFrame(data)

# Tokenization and preprocessing
max_words = 5000  # Vocabulary size
max_len = 100     # Maximum length of sequences

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(df['text'])
sequences = tokenizer.texts_to_sequences(df['text'])
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['sentiment'], test_size=0.2, random_state=42)


#Step 3: Build a Simple CNN Model
embedding_dim = 50  # Embedding dimensions for word vectors

model = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

#Step 4: Train the Model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
#Step 5: Evaluate the Model
# Evaluate on test data
y_pred = model.predict(X_test)
y_pred_classes = (y_pred > 0.5).astype(int).flatten()
accuracy = accuracy_score(y_test, y_pred_classes)
print(f"Test Accuracy: {accuracy:.2f}")



```

Que.5. How to implement a simple Seq2Seq model for machine translation using LSTM in Keras0


```python
import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, Dense
#from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras.layers import Input, Embedding, LSTM, Dense
#Step 2: Prepare the Dataset


# Example data
input_texts = [
    "Hello",
    "How are you?",
    "Good morning"
]
target_texts = [
    "Hola",
    "¿Cómo estás?",
    "Buenos días"
]

# Add start and end tokens to target sentences
target_texts = ['\t' + text + '\n' for text in target_texts]
#Step 3: Tokenize Texts
# Define tokenizers
input_tokenizer = Tokenizer(char_level=True)
target_tokenizer = Tokenizer(char_level=True)

# Fit tokenizers
input_tokenizer.fit_on_texts(input_texts)
target_tokenizer.fit_on_texts(target_texts)

# Convert texts to sequences
input_sequences = input_tokenizer.texts_to_sequences(input_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)

# Get vocabulary size
num_encoder_tokens = len(input_tokenizer.word_index) + 1
num_decoder_tokens = len(target_tokenizer.word_index) + 1

# Pad sequences
max_encoder_seq_length = max([len(seq) for seq in input_sequences])
max_decoder_seq_length = max([len(seq) for seq in target_sequences])

encoder_input_data = pad_sequences(input_sequences, maxlen=max_encoder_seq_length, padding='post')
decoder_input_data = pad_sequences([seq[:-1] for seq in target_sequences], maxlen=max_decoder_seq_length, padding='post')
decoder_target_data = pad_sequences([seq[1:] for seq in target_sequences], maxlen=max_decoder_seq_length, padding='post')




#Step 4: Build the Seq2Seq Model
from keras.models import Model
from keras.layers import Input, Embedding, LSTM, Dense

# Define model parameters
embedding_dim = 256  # Dimension of the embedding vectors
latent_dim = 256     # Latent dimensionality of the encoding space

# Encoder
encoder_inputs = Input(shape=(None,), name='encoder_inputs')
encoder_embedding = Embedding(input_dim=num_encoder_tokens,
                              output_dim=embedding_dim,
                              mask_zero=True,
                              name='encoder_embedding')(encoder_inputs)
encoder_lstm = LSTM(latent_dim,
                    return_state=True,
                    name='encoder_lstm')
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(None,), name='decoder_inputs')
decoder_embedding = Embedding(input_dim=num_decoder_tokens,
                              output_dim=embedding_dim,
                              mask_zero=True,
                              name='decoder_embedding')(decoder_inputs)
decoder_lstm = LSTM(latent_dim,
                    return_sequences=True,
                    return_state=True,
                    name='decoder_lstm')
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)

# Ensure Dense layer input shape is compatible
decoder_dense = Dense(num_decoder_tokens, activation='softmax', name='decoder_dense')
decoder_outputs = decoder_dense(decoder_outputs)

# Full Seq2Seq Model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
# Step 5: Compile and Train the Model

# Prepare target data with correct shape
decoder_target_data = np.expand_dims(decoder_target_data, -1)  # Shape: (num_samples, timesteps, 1)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

# Train the model
history = model.fit(
    [encoder_input_data, decoder_input_data],  # Input data
    decoder_target_data,                      # Target data
    batch_size=64,                            # Batch size
    epochs=50,                                # Number of epochs
    validation_split=0.2                      # Validation split
)

#step 6 _Decode interferance
from keras.models import Model
from keras.layers import Input, Embedding, LSTM, Dense

# Define dimensions (same as training)
latent_dim = 256  # Latent dimensionality
num_decoder_tokens = 1000  # Number of tokens in the target vocabulary

# Define inference input layers
decoder_state_input_h = Input(shape=(latent_dim,), name='decoder_state_input_h')
decoder_state_input_c = Input(shape=(latent_dim,), name='decoder_state_input_c')
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

# Decoder Embedding layer
decoder_inputs_inf = Input(shape=(None,), name='decoder_inputs_inf')  # Shape: (batch_size, timesteps)
decoder_embedding_layer = Embedding(
    input_dim=num_decoder_tokens, output_dim=latent_dim, mask_zero=True, name='decoder_embedding_inf'
)
decoder_embedding_inf = decoder_embedding_layer(decoder_inputs_inf)

# Decoder LSTM layer
decoder_lstm_layer = LSTM(latent_dim, return_sequences=True, return_state=True, name='decoder_lstm_inf')
decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm_layer(
    decoder_embedding_inf, initial_state=decoder_states_inputs
)

# Decoder Dense layer
decoder_dense_layer = Dense(num_decoder_tokens, activation='softmax', name='decoder_dense_inf')
decoder_outputs_inf = decoder_dense_layer(decoder_outputs_inf)

# Define decoder inference model
decoder_model = Model(
    [decoder_inputs_inf] + decoder_states_inputs,
    [decoder_outputs_inf] + [state_h_inf, state_c_inf]
)

# Display model summary
decoder_model.summary()


#Step 7: Perform Translation
def decode_sequence(input_seq):
    # Encode the input as state vectors.
    states_value = encoder_model.predict(input_seq)

    # Generate empty target sequence with only the start character.
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = target_tokenizer.word_index['\t']

    # Initialize variables
    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

        # Sample the token with highest probability
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = target_tokenizer.index_word.get(sampled_token_index, '')

        decoded_sentence += sampled_char

        # Exit condition: either hit max length or find the stop character.
        if (sampled_char == '\n' or
            len(decoded_sentence) > max_decoder_seq_length):
            stop_condition = True

        # Update the target sequence (of length 1).
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update states
        states_value = [h, c]

    return decoded_sentence

# Example translation
for seq_index in range(len(input_texts)):
    input_seq = encoder_input_data[seq_index:seq_index + 1]
    decoded_sentence = decode_sequence(input_seq)
    print(f"Input: {input_texts[seq_index]}")
    print(f"Output: {decoded_sentence}")
```

Que.7.2G How to apply data augmentation for text in NLP


```python



!pip install nlpaug # Install nlpaug first
import nltk
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet') # Download the wordnet dataset
import nlpaug.augmenter.word as naw # Then import


#Step-1. Synonym Replacement
from nltk.corpus import wordnet
import random

def synonym_replacement(text, n=1):
    words = text.split()
    new_words = words[:]
    random_word_list = list(set(words))
    random.shuffle(random_word_list)
    num_replaced = 0
    for random_word in random_word_list:
        synonyms = wordnet.synsets(random_word)
        if synonyms:
            synonym = random.choice(synonyms).lemmas()[0].name()
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:
            break
    return ' '.join(new_words)

text = "The quick brown fox jumps over the lazy dog."
print(synonym_replacement(text, n=2))


#step22. Back Translation

from transformers import MarianMTModel, MarianTokenizer

def back_translate(text, source_lang="en", pivot_lang="fr"):
    # Load translation model
    model_name = f"Helsinki-NLP/opus-mt-{source_lang}-{pivot_lang}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Translate to pivot language
    encoded = tokenizer(text, return_tensors="pt")
    translated = model.generate(**encoded)
    pivot_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    # Translate back to source language
    model_name = f"Helsinki-NLP/opus-mt-{pivot_lang}-{source_lang}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    encoded = tokenizer(pivot_text, return_tensors="pt")
    back_translated = model.generate(**encoded)
    return tokenizer.decode(back_translated[0], skip_special_tokens=True)

text = "The quick brown fox jumps over the lazy dog."
print(back_translate(text))


#step 3. Random Insertion
def random_insertion(text, n=1):
    words = text.split()
    for _ in range(n):
        new_word = random.choice(words)
        synonyms = wordnet.synsets(new_word)
        if synonyms:
            synonym = random.choice(synonyms).lemmas()[0].name()
            insert_pos = random.randint(0, len(words))
            words.insert(insert_pos, synonym)
    return ' '.join(words)

text = "The quick brown fox jumps over the lazy dog."
print(random_insertion(text, n=2))

#step 4. Random Deletion

def random_deletion(text, p=0.3):
    words = text.split()
    if len(words) == 1:
        return text
    new_words = [word for word in words if random.uniform(0, 1) > p]
    return ' '.join(new_words) if new_words else random.choice(words)

text = "The quick brown fox jumps over the lazy dog."
print(random_deletion(text, p=0.3))

#step 5. Word Shuffling
def word_shuffling(text):
    words = text.split()
    random.shuffle(words)
    return ' '.join(words)

text = "The quick brown fox jumps over the lazy dog."
print(word_shuffling(text))
#step 6. Contextual Augmentation
from transformers import pipeline

# Use fill-mask pipeline for contextual word substitution
unmasker = pipeline("fill-mask", model="bert-base-uncased")

def contextual_augmentation(text, mask_token="[MASK]"):
    words = text.split()
    random_idx = random.randint(0, len(words) - 1)
    words[random_idx] = mask_token
    masked_text = " ".join(words)
    suggestions = unmasker(masked_text)
    augmented_text = masked_text.replace(mask_token, suggestions[0]['token_str'])
    return augmented_text

text = "The quick brown fox jumps over the lazy dog."
print(contextual_augmentation(text))
#step.8.
def add_typo(text, n=1):
    words = list(text)
    for _ in range(n):
        idx = random.randint(0, len(words) - 2)
        words[idx], words[idx + 1] = words[idx + 1], words[idx]
    return ''.join(words)

text = "The quick brown fox jumps over the lazy dog."
print(add_typo(text, n=3))


#step9

# Synonym augmentation
aug = naw.SynonymAug(aug_src='wordnet')
augmented_text = aug.augment("The quick brown fox jumps over the lazy dog.")
print(augmented_text)
```

Que.8.How can you add an Attention Mechanism to a Seq2Seq model?


```python
import numpy as np
from keras.layers import Input, LSTM, Embedding, Dense, dot, Activation, Concatenate
from keras.models import Model

# Define some parameters
input_dim = 100  # Feature dimension of encoder input
latent_dim = 256  # Latent dimension of LSTM units
target_vocab_size = 50  # Size of the target vocabulary
embedding_dim = 50  # Embedding dimension for the decoder

# Encoder
encoder_inputs = Input(shape=(None, input_dim))  # Input shape: (timesteps, features)
encoder_lstm = LSTM(units=latent_dim, return_sequences=True, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_inputs)
encoder_states = [state_h, state_c]  # These will be passed to the decoder

# Attention Layer
def attention_layer(decoder_hidden_state, encoder_outputs):
    # Alignment scores: dot product of decoder hidden state and encoder outputs
    score = dot([decoder_hidden_state, encoder_outputs], axes=[2, 2])

    # Attention weights: softmax over the scores
    attention_weights = Activation('softmax')(score)

    # Context vector: weighted sum of encoder outputs
    context_vector = dot([attention_weights, encoder_outputs], axes=[2, 1])
    return context_vector, attention_weights

# Decoder inputs
decoder_inputs = Input(shape=(None,))  # Shape: (timesteps,)
decoder_embedding = Embedding(input_dim=target_vocab_size, output_dim=embedding_dim)(decoder_inputs)

# Decoder LSTM
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_lstm_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)

# Apply attention
context_vector, attention_weights = attention_layer(decoder_lstm_outputs, encoder_outputs)

# Concatenate context vector with LSTM outputs
decoder_combined_context = Concatenate(axis=-1)([context_vector, decoder_lstm_outputs])

# Output layer
decoder_dense = Dense(target_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_combined_context)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
model.summary()

# Example input data (adjust shapes to match expected model input)
# Encoder input shape: (batch_size, timesteps, input_dim)
encoder_input_data = np.random.rand(100, 10, input_dim)  # 100 samples, 10 timesteps, input_dim features

# Decoder input shape: (batch_size, timesteps)
decoder_input_data = np.random.randint(0, target_vocab_size, size=(100, 20))  # 100 samples, 20 timesteps

# Decoder target data shape: (batch_size, timesteps, 1)
decoder_target_data = np.random.randint(0, target_vocab_size, size=(100, 20, 1))  # 100 samples, 20 timesteps, 1 target class

# Train the model
model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=64,
    epochs=10,
    validation_split=0.2
)
```
