Que.3. How do you use TextBlob for performing text classification based on polarity


```python
!pip install textblob
from textblob import TextBlob

# Sample text
text = "I love the new phone! It has great features and amazing performance."

# Create a TextBlob object
blob = TextBlob(text)

# Get the polarity and subjectivity
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Print the sentiment analysis results
print(f"Polarity: {polarity}")
print(f"Subjectivity: {subjectivity}")

# Classify the sentiment based on polarity
if polarity > 0:
    sentiment = "Positive"
elif polarity == 0:
    sentiment = "Neutral"
else:
    sentiment = "Negative"

print(f"Sentiment Classification: {sentiment}")
```

Que.5.How can you calculate TF-IDF scores for a given text using Scikit-learn

```python
!pip install scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text (could be a single document or a list of documents)
documents = [
    "I love programming in Python.",
    "Python programming is fun and easy to learn.",
    "Natural language processing with Python is exciting."
]

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer on the documents and transform the documents into TF-IDF scores
tfidf_matrix = vectorizer.fit_transform(documents)

# Convert the TF-IDF matrix into an array and get the feature names (terms)
tfidf_array = tfidf_matrix.toarray()
terms = vectorizer.get_feature_names_out()

# Display the TF-IDF scores for each term in each document
for i, doc in enumerate(documents):
    print(f"Document {i+1}: {doc}")
    for j, term in enumerate(terms):
        print(f"  {term}: {tfidf_array[i][j]:.4f}")

```

Que.6. How do you create a custom text classifier using NLTK's Naive Bayes classifier

```python
!pip install nltk
import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download necessary NLTK data (if you haven't done this already)
nltk.download('punkt')
nltk.download('stopwords')

# Example dataset of labeled text
train_data = [
    ("I love programming in Python", "positive"),
    ("Python is a great programming language", "positive"),
    ("I hate bugs in my code", "negative"),
    ("Debugging is frustrating", "negative"),
    ("I enjoy learning new programming languages", "positive"),
    ("I am not a fan of Java", "negative")
]

# Preprocessing: Tokenize and remove stop words
stop_words = set(stopwords.words('english'))

def extract_features(text):
    # Tokenize the text
    words = word_tokenize(text.lower())
    # Remove stop words
    words = [word for word in words if word.isalpha() and word not in stop_words]

    # Create a feature dictionary: Each word is a feature
    features = {word: True for word in words}
    return features

# Prepare training data with extracted features
train_features = [(extract_features(text), label) for (text, label) in train_data]

# Train the Naive Bayes classifier
classifier = NaiveBayesClassifier.train(train_features)

# Test the classifier
test_data = [
    "I love writing code in Python",
    "I dislike debugging my Java code"
]

# Classify the test data
for text in test_data:
    features = extract_features(text)
    predicted_label = classifier.classify(features)
    print(f"Text: {text} => Predicted Label: {predicted_label}")

# Show the most informative features
classifier.show_most_informative_features(5)
```

Que.9.How can you create a simple RNN for text classification using Keras?

```python
!pip install tensorflow
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Sample data (text and labels)
texts = [
    "I love programming in Python",
    "I hate bugs in my code",
    "Python is awesome",
    "I enjoy solving problems",
    "Debugging is frustrating",
    "Java is not my favorite"
]
labels = [1, 0, 1, 1, 0, 0]  # 1 for positive, 0 for negative sentiment

# Parameters
max_words = 1000  # Maximum number of words in the vocabulary
max_len = 10      # Maximum length of each input sequence

# Tokenize the text data
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)

# Pad the sequences to ensure uniform input length
X = pad_sequences(X, maxlen=max_len)

# Convert to NumPy arrays
X = np.array(X)
labels = np.array(labels)

# Ensure labels are in the correct shape for binary classification
y_train = labels.reshape(-1, 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

# Check shapes of the data
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# Build the RNN model
model = Sequential()

# Add an embedding layer (to convert words into word vectors)
model.add(Embedding(input_dim=max_words, output_dim=50, input_length=max_len))

# Add an RNN layer (SimpleRNN)
model.add(SimpleRNN(units=64, activation='tanh'))

# Add a Dropout layer to prevent overfitting
model.add(Dropout(0.2))

# Add a Dense layer for classification (binary classification here)
model.add(Dense(1, activation='sigmoid'))  # For binary classification (0 or 1)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Make predictions
predictions = model.predict(X_test)
print(predictions)
```

Que.11.How do you implement GRU (Gated Recurrent Unit) for text classification

```python
!pip install tensorflow
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Sample data (text and labels)
texts = [
    "I love programming in Python",
    "I hate bugs in my code",
    "Python is awesome",
    "I enjoy solving problems",
    "Debugging is frustrating",
    "Java is not my favorite"
]
labels = [1, 0, 1, 1, 0, 0]  # 1 for positive, 0 for negative sentiment

# Parameters
max_words = 1000  # Maximum number of words in the vocabulary
max_len = 10      # Maximum length of each input sequence

# Tokenize the text data
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)

# Pad the sequences to ensure uniform input length
X = pad_sequences(X, maxlen=max_len)

# Convert to NumPy arrays if necessary
X = np.array(X)
labels = np.array(labels)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

# Ensure labels are in the correct shape
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Build the GRU model
model = Sequential()

# Add an embedding layer (to convert words into word vectors)
model.add(Embedding(input_dim=max_words, output_dim=50, input_length=max_len))

# Add a GRU layer
model.add(GRU(units=64, activation='tanh', return_sequences=False))

# Add a Dropout layer to prevent overfitting
model.add(Dropout(0.2))

# Add a Dense layer for classification (binary classification here)
model.add(Dense(1, activation='sigmoid'))  # For binary classification (0 or 1)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Make predictions
predictions = model.predict(X_test)
print(predictions)
```

Que.12. How do you implement a text generation model using LSTM with Keras?

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

# Sample text data (you can replace this with any large corpus)
text = """This is an example of text generation. Text generation uses machine learning to predict the next word or character based on the previous sequence."""

# Step 1: Preprocessing the data
# Create a list of unique characters in the text
chars = sorted(list(set(text)))
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for idx, char in enumerate(chars)}

# Step 2: Prepare input sequences and output labels
sequence_length = 40  # Length of each input sequence
sequences = []
next_chars = []

# Create sequences of 40 characters each
for i in range(len(text) - sequence_length):
    sequences.append(text[i:i + sequence_length])
    next_chars.append(text[i + sequence_length])

# Convert sequences to numerical data
X = np.zeros((len(sequences), sequence_length, len(chars)), dtype=bool)  # Changed np.bool to bool
y = np.zeros((len(sequences), len(chars)), dtype=bool)  # Changed np.bool to bool

for i, sequence in enumerate(sequences):
    for t, char in enumerate(sequence):
        X[i, t, char_to_idx[char]] = 1
    y[i, char_to_idx[next_chars[i]]] = 1

# Step 3: Build the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(sequence_length, len(chars)), return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(len(chars)))
model.add(Activation('softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001))  # Updated learning rate argument

# Step 4: Train the model
model.fit(X, y, batch_size=128, epochs=10)

# Step 5: Generate text using the trained model
def generate_text(seed, length=100):
    generated = seed
    for _ in range(length):
        # Prepare the input sequence
        X_pred = np.zeros((1, sequence_length, len(chars)))
        for t, char in enumerate(generated[-sequence_length:]):
            X_pred[0, t, char_to_idx[char]] = 1

        # Predict the next character
        prediction = model.predict(X_pred, verbose=0)
        next_char_idx = np.argmax(prediction)
        next_char = idx_to_char[next_char_idx]

        # Append the predicted character to the generated text
        generated += next_char
    return generated

# Test text generation with a seed text
seed_text = "This is an example"
generated_text = generate_text(seed_text, length=100)
print(generated_text)
```

Que.13.How do you implement a simple Bi-directional GRU for sequence labeling?


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, GRU, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Sample data: sequences of text and their corresponding labels
texts = [
    "I love programming",
    "Python is great",
    "I hate bugs",
    "Debugging is fun"
]
labels = [
    [1, 0, 0],  # Label for each word (I, love, programming)
    [1, 0, 0],  # Python, is, great
    [0, 1, 0],  # I, hate, bugs
    [0, 0, 1]   # Debugging, is, fun
]

# Define parameters
max_words = 1000  # Maximum number of words in the vocabulary
max_len = 10      # Maximum length of each input sequence
embedding_dim = 50
num_classes = 3   # Number of classes (labels)

# Tokenize the text data
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, maxlen=max_len)

# One-hot encode the labels for sequence labeling (for each word)
y = np.array(labels)  # Convert list of lists into a numpy array
y = np.expand_dims(y, axis=-1)  # Add an extra dimension to the labels (if needed)
y = np.eye(num_classes)[y.squeeze()]  # One-hot encode the labels, remove unnecessary dimension

# Ensure the target shape matches the output shape of the model: (batch_size, sequence_length, num_classes)
y = np.pad(y, ((0, 0), (0, max_len - y.shape[1]), (0, 0)), 'constant')  # Pad the labels to match sequence length

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Build the Bi-directional GRU model
model = Sequential()

# Add an embedding layer (to convert words into word vectors)
model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len))

# Add a Bi-directional GRU layer
model.add(Bidirectional(GRU(64, return_sequences=True)))  # Using GRU with return_sequences=True for sequence labeling

# Add Dropout to prevent overfitting
model.add(Dropout(0.5))

# Add a Dense layer for classification (per timestep)
model.add(Dense(num_classes, activation='softmax'))  # 3 classes for labeling each word

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Make predictions on new data
predictions = model.predict(X_test)
print("Predictions: ", np.argmax(predictions, axis=-1))  # Predicted labels
```
